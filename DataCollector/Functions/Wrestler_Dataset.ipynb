#   Adding Competitors to Dataset

```python
all_names = set()
for competitors in dataset['competitors']:
    names = competitors.split(', ')
    all_names.update(names)

# Create a new DataFrame with the competitor names
filtered_dataset = pd.DataFrame({'wrestler_name': list(all_names)})
```

#   Adding to Amount of Matches Involved In

```python
competitor_counts = {}
for name in all_names:
    if name in competitor_counts:
        competitor_counts[name] += 1
    else:
        competitor_counts[name] = 1

# Create DataFrame for competitor counts
filtered_dataset = pd.DataFrame({'competitor_name': list(competitor_counts.keys()), 'count': list(competitor_counts.values())})
```
