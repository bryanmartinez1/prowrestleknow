#   Adding Competitors to Dataset

```python
all_names = set()
for competitors in dataset['competitors']:
    names = competitors.split(', ')
    all_names.update(names)

# Create a new DataFrame with the competitor names
filtered_dataset = pd.DataFrame({'wrestler_name': list(all_names)})
filtered_dataset.head()
```

#   Adding to Amount of Matches Involved In

```python
counts = []
for wrestler in filtered_dataset["wrestler_name"]:
    count = 0
    for matches in dataset['competitors']:
        names = matches.split(', ')
        if (wrestler in names):
            count+=1
    counts.append(count)
filtered_dataset['matches'] = counts

filtered_dataset.head(5)
```

#   Adding to Amount of Wins Involved In

```python
counts = []
for wrestler in filtered_dataset["wrestler_name"]:
    count = 0
    for matches in dataset['winners']:
        names = matches.split(', ')
        if (wrestler in names):
            count+=1
    counts.append(count)
filtered_dataset['wins'] = counts

filtered_dataset.head(30)
```
